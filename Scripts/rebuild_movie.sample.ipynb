```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_orig = np.load('../Data/2frames.npz')
data_f0=np.load('../Data/createdFile0.npz')
data_f1=np.load('../Data/createdFile1.npz')
```

```python
oldimg_f0=data_orig['data'][:,:,0]
oldimg_f1=data_orig['data'][:,:,1]
newimg_f0=data_f0['newimg']
newimg_f1=data_f1['newimg']
```

```python
newimg_f0=newimg_f0.reshape((1024,1000))
newimg_f1=newimg_f1.reshape((1024,1000))
```

```python
plt.subplot(131)
plt.imshow(oldimg_f0,cmap=plt.get_cmap('gray'))
plt.subplot(132)
plt.imshow(newimg_f0,cmap=plt.get_cmap('gray'))
plt.subplot(133)
plt.imshow(oldimg_f1 - newimg_f0,cmap=plt.get_cmap('gray'))
```

```python
plt.subplot(131)
plt.imshow(oldimg_f1,cmap=plt.get_cmap('gray'))
plt.subplot(132)
plt.imshow(newimg_f1,cmap=plt.get_cmap('gray'))
plt.subplot(133)
plt.imshow(oldimg_f1 - newimg_f1,cmap=plt.get_cmap('gray'))
```

```python
plt.imsave('../Data/old_frame0.tif',oldimg_f0,cmap=plt.get_cmap('gray'))
plt.imsave('../Data/old_frame1.tif',oldimg_f1,cmap=plt.get_cmap('gray'))

plt.imsave('../Data/new_frame0.tif',newimg_f0,cmap=plt.get_cmap('gray'))
plt.imsave('../Data/new_frame1.tif',newimg_f1,cmap=plt.get_cmap('gray'))
```
